Projeto onde vou analisar a base de dados e mandar um email para o dono das lojas mostrando a quantidade de produtos vendidas e quanto faturou no mês.
[ATENÇÃO] As coordenadas do click do mouse pode não funcionar em outros monitores pois as coordenadas estão de acordo com o meu monitor.

```python
import pandas as pd

# Importando a base de dados
tabela = pd.read_excel('Vendas - Dez.xlsx')

# Exibindo a base de dados na tela
display(tabela)
```

```python
# Calcular quantidade de produtos vendidos
qnt_produtos_vendidos = tabela['Quantidade'].sum()

# Calcular faturamento
faturamento = tabela['Valor Final'].sum()
```

```python
import pyautogui as gui
import pyperclip as clip
import time

# Enviar email automaticamente
gui.hotkey('ctrl', 't') # Abre uma nova guia
time.sleep(1) # Espera a guia carregar
clip.copy('https://mail.google.com') # Copia o link do gmail
gui.hotkey('ctrl', 'v') # Cola o link no navegador
time.sleep(1) # Espera colar no navegador
gui.press('enter') # Pressiona a tecla enter
time.sleep(4) # Espera
gui.click(x=105, y=197) # Clica para enviar o email
time.sleep(2) # Espera
email = 'vinibrandaogim@hotmail.com' # Endereço de email que vai receber
gui.write(email) # Escreve o email
time.sleep(3) # Espera  
gui.press('tab') # Confirma o endereço de email que vai receber o email
time.sleep(2) # Espera 
gui.press('tab') # Pula para o assunto do email
clip.copy('Relatório de vendas') # Assunto do email
gui.hotkey('ctrl', 'v') # Escreve o assunto
gui.press('tab') # Pula para o texto do email
time.sleep(2) # Espera
clip.copy(f'''Olá chefe,

    O faturamento desse mês foi de R$ {faturamento:,.2f} e a quantidade de produtos vendidos foi de {qnt_produtos_vendidos:,}

Abraços''') # Copia o texto
gui.hotkey('ctrl', 'v') # Cola o texto
time.sleep(2) # Espera
gui.press('tab') # Pula para enviar o email
gui.press('enter') # Pressiona a tecla enter
```
